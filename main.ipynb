```python
import MySQLdb
import tkinter
import pandas as pd
%matplotlib inline

db = MySQLdb.connect("localhost", "root", "2ZombiesEatBrains?")
cursor = db.cursor()

df_cols = [
    "ID", "Name", "Category", "Stock Available", "Selling Price", "Description"
]

cursor.execute("SELECT * FROM sprint_datavault.products")
db_data = cursor.fetchall()

db.close()


products_df = pd.DataFrame(db_data, columns=df_cols).set_index("ID")

products_df.groupby(["Category"]).size().plot(y="Stock Available", kind="bar", title="Number of Items per Category")
products_df.groupby(["Category"]).sum().plot(y="Stock Available", kind="bar", title="Total Number of Products per Category")
products_df.groupby(["Category"]).mean().plot(y="Stock Available", kind="bar", title="Average Price of Products in Category")
```
